```python
# Install required libraries (if not already installed)
!pip install textstat

import nltk
import textstat
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

# Function to calculate Flesch-Kincaid Grade Level (FKGL) manually
def flesch_kincaid_grade(sentence):
    words = word_tokenize(sentence)  # Tokenize words
    num_words = len(words)  # Count words
    num_sentences = len(sent_tokenize(sentence))  # Count sentences
    num_syllables = sum([textstat.syllable_count(word) for word in words])  # Count syllables

    # Calculate ASL (Average Sentence Length) and ASW (Average Syllables per Word)
    ASL = num_words / num_sentences if num_sentences > 0 else 0
    ASW = num_syllables / num_words if num_words > 0 else 0

    # Apply the FKGL formula
    FKGL = (0.39 * ASL) + (11.8 * ASW) - 15.59
    return round(FKGL, 2)  # Return rounded score

# Compute FKGL for each sentence
df['Flesch-Kincaid Score'] = df['sentences'].apply(flesch_kincaid_grade)

# Categorize sentences into Easy, Medium, Hard
def categorize_fkgl(score):
    if score <= 5:
        return "Easy"
    elif 6 <= score <= 8:
        return "Medium"
    else:
        return "Hard"

df
```

```python
# Install required libraries
!pip install --no-cache-dir pandas openpyxl textstat nltk

import pandas as pd
import textstat
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# ✅ Fix: Download all required NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')

# ✅ Load the dataset
df = pd.read_excel("sentences.xlsx")

if 'sentences' not in df.columns:
    raise ValueError("Column 'sentences' not found in dataset. Check your file.")

# ✅ Define FKGL function
def flesch_kincaid_grade(sentence):
    words = word_tokenize(str(sentence))  # Ensure text format
    num_words = len(words)
    num_sentences = len(sent_tokenize(str(sentence)))
    num_syllables = sum([textstat.syllable_count(word) for word in words])

    if num_sentences == 0 or num_words == 0:  # Avoid division by zero
        return 0

    ASL = num_words / num_sentences
    ASW = num_syllables / num_words
    FKGL = (0.39 * ASL) + (11.8 * ASW) - 15.59

    return round(FKGL, 2)

# ✅ Apply FKGL score
df['Flesch-Kincaid Score'] = df['sentences'].apply(flesch_kincaid_grade)

# ✅ Categorize sentences
def categorize_fkgl(score):
    if score <= 5:
        return "Easy"
    elif 6 <= score <= 8:
        return "Medium"
    else:
        return "Hard"

df['Category'] = df['Flesch-Kincaid Score'].apply(categorize_fkgl)

# ✅ Save processed dataset
df.to_excel("processed_sentences.xlsx", index=False)
print("✅ Processed dataset saved as 'processed_sentences.xlsx'")
```
